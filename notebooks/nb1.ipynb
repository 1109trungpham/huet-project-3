```python
from pathlib import Path

# Tạo đối tượng Path trỏ đến file
path = Path("../data/TripAdvisor_Data_Cleaned_Hotel_English_Hue.json")

# Biến nó thành đường dẫn tuyệt đối
absolute_path = path.resolve()

print(f"Đường dẫn tuyệt đối: {absolute_path}")
```

```python
# Kiểm tra xem file có thực sự tồn tại không trước khi đọc
import pandas as pd

if absolute_path.exists():
    df = pd.read_json(absolute_path)
    print("Đã tải dữ liệu thành công!")
else:
    print("Cảnh báo: File vẫn không tồn tại tại vị trí này.")
```
